```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
from scipy import special

py.offline.init_notebook_mode(connected=True)

from plotly import tools

FOURIER = "Fourier"
FREQ = "Frequency"
TIME = 'X_Value'
VOLTAGE = 'Voltage - Dev1_ai0'
```

```python

FILENAME = '../data/test2hz.xlsx'
sample_rate = 1000 #1/deltaX in csv file
STOP = 10 #duration of measurement (in seconds)


data = pd.read_excel(FILENAME, header=7, usecols=[0, 1])


data[FOURIER] = 2/(sample_rate)*pd.Series(np.abs(np.fft.fft(data[VOLTAGE])))*STOP**-1
data[FREQ] = pd.Series(sample_rate*data[TIME])*STOP**-1

s = go.Scatter( x=data[TIME],y=data[VOLTAGE],mode='markers')
s2 = go.Scatter(x=data[FREQ],y=data[FOURIER], mode='markers')

fig = tools.make_subplots(rows=1,cols=2)
fig.append_trace(s,1,1)
fig.append_trace(s2,1,2)

py.offline.iplot(fig)
print(data)
```

get Signal to Noise ratio:
# NOT AUTOMATIC, MANUALLY DETECT AND REMOVE SIGNAL FREQUENCIES


```python
#remove signal peaks

PEAK_INDICES = [20,9980]
signal_sum=0

data2 = data
for i in PEAK_INDICES:
    signal_sum+=data2.at[i,FOURIER]
    data2 = data2.drop([i])

noise = data2[FOURIER].sum()

stn = signal_sum/noise

print("noise,signal,ratio:")
print(noise,signal_sum,stn)
```

plot noise data

```python
s3=go.Scatter(x=data2[FREQ],y=data2[FOURIER],mode='markers')
py.offline.iplot(go.Figure([s3]))
print(data2)
```
