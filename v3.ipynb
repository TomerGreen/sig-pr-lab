```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
from scipy import special

py.offline.init_notebook_mode(connected=True)

from plotly import tools
TIME = "Time"
VOLTAGE = "Voltage"
FOURIER = "Fourier"
FREQ = "Frequency"
```

sample rate should be at least 2 times max frequency.

Frequencies to consider are up to half the sample_rate.

```python
START=0
STOP=10
STEP=0.01

def update_plot(freq1,amp1,freq2,amp2,sample_rate):
    
    time = pd.Series(np.arange(START,STOP,1/sample_rate))
    
    
    
    data = pd.DataFrame()
    data[TIME] = time
    data[VOLTAGE] = amp1*np.sin(2*np.pi*freq1*time)
    # add another sin wave
    data[VOLTAGE] += amp2*np.sin(2*np.pi*freq2*time)
    
    
    data[FOURIER] = 2*sample_rate**-1*pd.Series(np.abs(np.fft.fft(data[VOLTAGE])))*STOP**-1
    data[FREQ] = pd.Series(sample_rate*(time))*STOP**-1
    


    s = go.Scatter( x=data[TIME],y=data[VOLTAGE],mode='markers')
    s2 = go.Scatter(x=data[FREQ],y=data[FOURIER], mode='markers')
    
    fig = tools.make_subplots(rows=1,cols=2)
    fig.append_trace(s,1,1)
    fig.append_trace(s2,1,2)
    
    py.offline.iplot(fig)
    print(data)
    
freq1 = widgets.FloatSlider(min=0,max=300, step=1 ,value=0,description="Freq1")
freq2 = widgets.FloatSlider(min=0,max=300, step=1 ,value=1,description="Freq2")
amp1 = widgets.FloatSlider(min=0,max=300, step=1 ,value=0,description="Amp1")
amp2 = widgets.FloatSlider(min=0,max=300, step=1 ,value=1,description="Amp2")
sample_rate = widgets.FloatSlider(min=1,max=1000, step=10 ,value=100,description="Sample rate")
widgets.interactive(update_plot,freq1=freq1, amp1= amp1,amp2= amp2, freq2=freq2, sample_rate=sample_rate)
```

```python
#remove signal peaks

PEAK_INDICES = [1,29,71,99]
signal_sum=0

data2 = data
for i in PEAK_INDICES:
    signal_sum+=data2.at[i,FOURIER]
    data2 = data2.drop([i])

noise = data2[FOURIER].sum()

stn = signal_sum/noise

print("noise,signal,ratio:")
print(noise,signal_sum,stn)
```
